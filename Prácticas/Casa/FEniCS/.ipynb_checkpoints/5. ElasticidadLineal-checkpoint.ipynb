## __Ecuación Diferencial Parcial - _PDE___

<div align="justify">
    Las ecuaciones gobernantes para pequeñas deformaciones elásticas de un cuerpo $\Omega$ se puede resumir de la siguiente manera:
    $$
    \begin{equation}
        -\nabla \cdot \sigma = f \, \,\,\, \text{en} \, \Omega , \\
        \sigma = \lambda tr (\varepsilon) I + 2\mu \varepsilon, \\
        \varepsilon = \frac{1}{2} \left(\nabla u + \left(\nabla u \right)^T \right)
        \tag{1}
        \label{pde}
    \end{equation}
    $$
    De la Ecuación \ref{pde}, $\sigma$ es el tensor de esfuerzos, $f$ es la fuerza distribuida por unidad de volumen, $\lambda$ y $\mu$ son los parámetros de elasticidad de <i>Lamè</i> para el material en $\Omega$, $I$ es el tensor identidad, $tr$ es el operador de rastro sobre el tensor, $\varepsilon$ se trata de la variación de la deformación del tensor (gradiente simétrico), y $u$ es el vector de campo de desplazamiento. Lo anterior asumiendo condiciones elásticas de carácter isotrópico.
    $$
    \begin{equation}
        \sigma = \lambda \left(\nabla \cdot u \right) I + \mu \left(\nabla u + \left(\nabla u \right)^T \right)
        \tag{2}
        \label{sigma}
    \end{equation}
    $$
</div>

## __Formulación Variacional__

<div align="justify">
    La formulación variacional consiste en la formación del producto interno de la Ecuación \ref{pde} y la función del vector de prueba, como se muestra en la Ecuación \ref{var}.
    $$
    \begin{equation}
        -\int _{\Omega} \left(\nabla \cdot \sigma \right)\cdot v \, dx = \int _{\Omega} f\cdot v \, dx
        \tag{3}
        \label{var}
    \end{equation}
    $$
    Integrando por partes...
    $$
    \begin{equation}
        - \int _{\Omega} \left(\nabla \cdot \sigma \right) \cdot v \, dx = \int _{\Omega} \sigma : \nabla v \, dx - \int _{\partial \Omega} \left(\sigma \cdot n \right) \cdot v \, ds
        \tag{4}
        \label{int}
    \end{equation}
    $$
    De la Ecuación \ref{int}, el operador de dos puntos $\left(\sigma : \nabla v \right)$ se trata del producto interno entre tensores y $n$ es el vector unitario normal hacia afuera sobre la condición de frontera. La cantidad $\sigma \cdot n$ es conocida como la <i>tracción</i> o el vector de esfuerzo sobre la frontera (de modo que: $\sigma \cdot n = T$), y se preescribe como una condición de frontera. Se asume que el valor del desplazamiento se da como una condición tipo Dirichlet.
    $$
    \begin{equation}
        \int _{\Omega} \sigma : \nabla v \, dx = \int _{\Omega} f \cdot v \, dx + \int _{\partial \Omega _T} T \cdot v \, ds
        \tag{5}
        \label{sigmados}
    \end{equation}
    $$
    Finalmente, para relacionar la forma variacional lineal $\left(a(u,v) = L(v) \right)$ con las Ecuaciones \ref{sigmados} y \ref{sigma}, se tiene que:
    $$
    \begin{equation}
        a(u,v) = \int _{\Omega} \sigma : \nabla v \, dx \\
        a(u,v) = \int _{\Omega} \sigma : \varepsilon (v) \, dx \\
        \varepsilon(v) = \frac{1}{2} \left(\nabla v + \left(\nabla v \right)^T \right) \\
        L(v) = \int _{\Omega} f \cdot v \, dx + \int _{\partial \Omega _T} T \cdot v \, ds
        \tag{6}
        \label{var2}
    \end{equation}
    $$
</div>

### Problema de prueba

<div align="justify">
    Se modelará una viga empotrada en un extremo y deformada bajo su propio peso en 3D.
</div>

* $f = (0,0,-\rho g)$
* Sección transversal de tipo cuadrada y ancho $W$.
* $u = u_D = (0,0,0)$ en $x=0$
* $T=0$

```python
%matplotlib inline
from fenics import *
from ufl import nabla_div
import numpy as np

#Datos generales
L = 1
W = 0.2
mu = 1
rho = 1
delta = W/L
gamma = 0.4*delta**2
beta = 1.25
lambda_ = beta
g = gamma

#Mallado y función espacial
mesh = BoxMesh(Point(0,0,0), Point(L, W, W), 10,3,3)
V = VectorFunctionSpace(mesh, 'P', 1)

#Condición de frontera
tol = 1E-14

def empotrado(x, on_boundary):
    return on_boundary and x[0] < tol

bc = DirichletBC(V, Constant((0,0,0)), empotrado)

#Esfuerzo y deformación

def epsilon(u):
    return 0.5*(nabla_grad(u) + nabla_grad(u).T)
    #return sym(nabla_grad(u))

def sigma(u):
    return lambda_*nabla_div(u)*Identity(d) + 2*mu*epsilon(u)

#Variables del problema variacional
u = TrialFunction(V)
d = u.geometric_dimension()    #Dimensión espacial
v = TestFunction(V)
f = Constant((0,0,-rho*g))
T = Constant((0,0,0))
a = inner(sigma(u), epsilon(v))*dx
L = dot(f, v)*dx + dot(T, v)*ds

# Solución
u = Function(V)
solve(a == L, u, bc)

# Graficar solución
#plot(u, title='Displacement', mode='displacement')

# Esfuerzo
s = sigma(u) - (1./3)*tr(sigma(u))*Identity(d)
von_Mises = sqrt(3./2*inner(s, s))
V = FunctionSpace(mesh, 'P', 1)
von_Mises = project(von_Mises, V)
#plot(von_Mises, title='Stress intensity')

# CDesplazamientos
u_magnitude = sqrt(dot(u, u))
u_magnitude = project(u_magnitude, V)
# plot(u_magnitude, 'Displacement magnitude')
print('min/max u:',
      u_magnitude.vector().get_local().min(),
      u_magnitude.vector().get_local().max())

# Guardar solución en formato VTK
File('elasticity/displacement.pvd') << u
File('elasticity/von_mises.pvd') << von_Mises
File('elasticity/magnitude.pvd') << u_magnitude

# Hold plot
# interactive()
```
